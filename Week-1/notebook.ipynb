```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge

from sklearn.metrics import mean_squared_error
```

```python
df_train = pd.read_parquet('./dataset/yellow_tripdata_2023-01.parquet')
df_val = pd.read_parquet('./dataset/yellow_tripdata_2023-02.parquet')
```

**Question 1.**
There are 19 columns.

```python
df_train.tpep_dropoff_datetime = pd.to_datetime(df_train.tpep_dropoff_datetime)
df_train.tpep_pickup_datetime = pd.to_datetime(df_train.tpep_pickup_datetime)

df_val.tpep_dropoff_datetime = pd.to_datetime(df_val.tpep_dropoff_datetime)
df_val.tpep_pickup_datetime = pd.to_datetime(df_val.tpep_pickup_datetime)
```

```python
df_train['duration'] = df_train.tpep_dropoff_datetime - df_train.tpep_pickup_datetime
df_train.duration = df_train.duration.apply(lambda td: td.total_seconds() / 60)

df_val['duration'] = df_val.tpep_dropoff_datetime - df_train.tpep_pickup_datetime
df_val.duration = df_val.duration.apply(lambda td: td.total_seconds() / 60)
```

```python
df_train.duration.describe()
```

**Question 2.** Standard deviation is 42.59

```python
((df_train.duration >= 1) & (df_train.duration <= 60)).mean()
```

```python
df_train = df_train[(df_train['duration'] > 1) & (df_train['duration'] < 60) ]
df_val = df_train[(df_train['duration'] > 1) & (df_train['duration'] < 60) ]
```

**Question 3.** Fraction of the records left after outliers are dropped is 98.12%

```python
categorical = ['PULocationID', 'DOLocationID']
numerical = ['trip_distance']
```

```python
df_train[categorical] = df_train[categorical].astype(str)
```

**Question 4.** Dimensionality of this matrix is 2.

```python
train_dicts = df_train[categorical + numerical].to_dict(orient='records')
```

```python
dv = DictVectorizer()
X_train = dv.fit_transform(train_dicts)

val_dicts = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dicts)
```

```python
X_train
```

```python
dv.feature_names_
```

```python
target = 'duration'
y_train = df_train[target].values
y_val = df_val[target].values
```

```python
y_train
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_val)
```

```python
sns.distplot(y_pred, label='predictions')
sns.distplot(y_train, label='actual')

plt.legend()
```

```python
mse = mean_squared_error(y_train, y_pred)
rmse = np.sqrt(mse)
```
